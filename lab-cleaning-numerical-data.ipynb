```python
#Felipe's feedback

# nulls method isn't doing anything, it's just to show that you can have multiple in a class

class preprocessing():
    
    def funk(self, data):
        self.categoricals = data.select_dtypes('object')
        self.discrete = data.select_dtypes('int64')
        self.continuous = data.select_dtypes('float64')
        return categoricals, discrete, continuous
    
    def nulls(self, data):
        is_na = categoricals.isna()
        return is_na
```
